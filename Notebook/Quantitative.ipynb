```python
import talib as ta
import pandas as pd

qnt['RSI']=ta.RSI(qnt['Close'],timeperiod=10)
```

```python
qnt['Signals']='Neutral'
qnt.loc[qnt['RSI']>70, 'Signals']='Buy'
qnt.loc[qnt['RSI']<30, 'Signals']='Sell'
```

```python
qnt[qnt['Signals']=='Sell']
```

```python
qnt['MACD_line'],qnt['MACD_sginal'],qnt['MACD_hist']=ta.MACD(qnt['Close'], signalperiod=10)
qnt
```

```python
qnt.isna().sum()
```

```python
rsi_mean,macd_mean=qnt['RSI'].mean(),qnt['MACD_line'].mean()
rsi_mean,macd_mean
```

```python
#filled the nan to mean of each column
qnt.loc[qnt['RSI'].isna(),'RSI']=rsi_mean
qnt.loc[qnt['MACD_line'].isna(),'MACD_line']=macd_mean

qnt
```

```python
import matplotlib.pyplot as plt

data_size=100
fig,ax=plt.subplots(4,1,figsize=(20,20))

ax[1].plot(qnt['RSI'].head(data_size))
ax[1].set_title('RSI')
ax[2].plot(qnt['MACD_line'].head(data_size))
ax[2].plot(qnt['MACD_sginal'].head(data_size))
ax[2].set_title('MACD_sginal and MACD_line')
ax[2].legend(loc='upper left', fontsize=12)
ax[3].plot(qnt['MACD_hist'].head(data_size))
ax[3].set_title('MACD_hist')
ax[0].plot(qnt['Close'].head(data_size))
ax[0].set_title('Close')

```
