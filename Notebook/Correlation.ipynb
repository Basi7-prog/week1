```python
resampled_numeric = publishers_names[['polarity', 'subjectivity']].resample('D').mean().dropna()
resampled_non_numeric = publishers_names[['headline', 'url','stock']].resample('D').first().dropna()
resampled_non_numeric
```

```python
import pandas as pd
new_df=pd.concat([resampled_non_numeric,resampled_numeric], axis=1)
new_df=new_df.dropna()
new_df
```

```python
aapl=Stocks.read_Stock_Data('AAPL')
Stocks.stock_movement_percental(aapl)
new_aapl=aapl[aapl.index.isin(new_df.index)]
new_aapl=new_aapl.dropna()
print(len(new_aapl), len(new_df))
new_aapl.
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots(2,1,figsize=(20,20))


ax[0].plot(new_aapl['Pct_Change'], label='stock movement')
ax[0].plot(new_df['polarity'], label='polarity')
ax[0].legend(fontsize=32)
# ax[0].legend(fontsize=32)
plt.show()
```

```python
polarity_sm=pd.concat([new_aapl['Pct_Change'],new_df['polarity']/100], axis=1)

polarity_sm=polarity_sm.dropna()
polarity_sm.loc[polarity_sm['Pct_Change']>0,'Pct_Change']=1
polarity_sm.loc[polarity_sm['Pct_Change']<0,'Pct_Change']=-1
polarity_sm.loc[polarity_sm['Pct_Change']==0,'Pct_Change']=0

polarity_sm.loc[polarity_sm['polarity']>0,'polarity']=1
polarity_sm.loc[polarity_sm['polarity']<0,'polarity']=-1
polarity_sm.loc[polarity_sm['polarity']==0,'polarity']=0


len(polarity_sm)
```

```python
#stock affected the day article is published
same_day=len(polarity_sm[polarity_sm['Pct_Change']==polarity_sm['polarity']])/len(polarity_sm)*100
same_day
nxt=polarity_sm.copy()
nxt['Pct_Change']=nxt['Pct_Change'].shift(1)
nxt=nxt.dropna()
nxt_day=len(nxt[nxt['Pct_Change']==nxt['polarity']])/len(nxt)*100
nxt_day,same_day
```

```python
from scipy.stats import pearsonr, spearmanr

polarity_sm2=pd.concat([new_aapl['Pct_Change'],new_df['polarity']/100], axis=1)
polarity_sm2=polarity_sm2.dropna()
# Pearson correlation
pearson_corr, p_value_pearson = pearsonr(polarity_sm2['Pct_Change'],polarity_sm2['polarity'])
spearmanr_corr, p_value_spearmanr = spearmanr(polarity_sm2['Pct_Change'],polarity_sm2['polarity'])
print(f"Pearson correlation: {pearson_corr}, p-value: {p_value_pearson}")
print(f"Spearman correlation: {spearmanr_corr}, p-value: {p_value_spearmanr}")
polarity_sm2
```

```python
import seaborn as sns

# Scatter plot with regression line
plt.figure(figsize=(10,10))
sns.regplot(x='polarity', y='Pct_Change', data=polarity_sm2)

# plt.plot(polarity_sm2)
# plt.xticks(polarity_sm2['polarity'])
# plt.yticks(polarity_sm2['Pct_Change'])
```
